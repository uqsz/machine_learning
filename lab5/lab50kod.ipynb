```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN, SpectralClustering
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Wczytanie danych
circles = pd.read_csv('data/circles.txt', header=None)
moons = pd.read_csv('data/moons.txt', header=None)
ellipses = pd.read_csv('data/ellipses.txt', header=None)
blobs = pd.read_csv('data/blobs.txt', header=None)
```

```python
def plot_clusters(data, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(data[0], data[1], c=labels, cmap='viridis')
    plt.title(title)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()
```

```python
# Parametry dla k-means
n_clusters_values = [2, 3, 4]
gamma_values = [0.1, 1, 10]

accuracies_kmeans_circles = []
accuracies_kmeans_moons = []
accuracies_kmeans_ellipses = []
accuracies_kmeans_blobs = []
```

```python
# Eksperymenty na zbiorze "circles" dla KMeans
for n_clusters in n_clusters_values:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(circles.iloc[:, :2])
    accuracy = accuracy_score(circles.iloc[:, 2], labels)
    accuracies_kmeans_circles.append(accuracy)

# Eksperymenty na zbiorze "moons" dla KMeans
for n_clusters in n_clusters_values:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(moons.iloc[:, :2])
    accuracy = accuracy_score(moons.iloc[:, 2], labels)
    accuracies_kmeans_moons.append(accuracy)

# Eksperymenty na zbiorze "ellipses" dla KMeans
for n_clusters in n_clusters_values:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(ellipses.iloc[:, :2])
    accuracy = accuracy_score(ellipses.iloc[:, 2], labels)
    accuracies_kmeans_ellipses.append(accuracy)

# Eksperymenty na zbiorze "blobs" dla KMeans
for n_clusters in n_clusters_values:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(blobs.iloc[:, :2])
    accuracy = accuracy_score(blobs.iloc[:, 2], labels)
    accuracies_kmeans_blobs.append(accuracy)

print("Accuracy for KMeans:")
print("Circles Dataset:")
for i, acc in enumerate(accuracies_kmeans_circles):
    print(f"N Clusters: {n_clusters_values[i]}, Accuracy: {acc:.2f}")

print("\nMoons Dataset:")
for i, acc in enumerate(accuracies_kmeans_moons):
    print(f"N Clusters: {n_clusters_values[i]}, Accuracy: {acc:.2f}")

print("\nEllipses Dataset:")
for i, acc in enumerate(accuracies_kmeans_ellipses):
    print(f"N Clusters: {n_clusters_values[i]}, Accuracy: {acc:.2f}")

print("\nBlobs Dataset:")
for i, acc in enumerate(accuracies_kmeans_blobs):
    print(f"N Clusters: {n_clusters_values[i]}, Accuracy: {acc:.2f}")
```

```python
# Parametry dla DBSCAN
eps_values = [0.1, 0.2, 0.3]
min_samples_values = [5, 10, 15]

accuracies_dbscan_circles = []
accuracies_dbscan_moons = []
accuracies_dbscan_ellipses = []
accuracies_dbscan_blobs = []
```

```python

# Eksperymenty na zbiorze "circles" dla DBSCAN
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(circles.iloc[:, :2])
        if len(set(labels)) > 1:
            accuracy = accuracy_score(circles.iloc[:, 2], labels)
            accuracies_dbscan_circles.append(accuracy)

# Eksperymenty na zbiorze "moons" dla DBSCAN
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(moons.iloc[:, :2])
        if len(set(labels)) > 1:
            accuracy = accuracy_score(moons.iloc[:, 2], labels)
            accuracies_dbscan_moons.append(accuracy)

# Eksperymenty na zbiorze "ellipses" dla DBSCAN
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(ellipses.iloc[:, :2])
        if len(set(labels)) > 1:
            accuracy = accuracy_score(ellipses.iloc[:, 2], labels)
            accuracies_dbscan_ellipses.append(accuracy)

# Eksperymenty na zbiorze "blobs" dla DBSCAN
for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(blobs.iloc[:, :2])
        if len(set(labels)) > 1:
            accuracy = accuracy_score(blobs.iloc[:, 2], labels)
            accuracies_dbscan_blobs.append(accuracy)

print("\n\nAccuracy for DBSCAN:")
print("Circles Dataset:")
for i, acc in enumerate(accuracies_dbscan_circles):
    print(f"Eps: {eps_values[i // len(min_samples_values)]}, Min Samples: {min_samples_values[i % len(min_samples_values)]}, Accuracy: {acc:.2f}")

print("\nMoons Dataset:")
for i, acc in enumerate(accuracies_dbscan_moons):
    print(f"Eps: {eps_values[i // len(min_samples_values)]}, Min Samples: {min_samples_values[i % len(min_samples_values)]}, Accuracy: {acc:.2f}")

print("\nEllipses Dataset:")
for i, acc in enumerate(accuracies_dbscan_ellipses):
    print(f"Eps: {eps_values[i // len(min_samples_values)]}, Min Samples: {min_samples_values[i % len(min_samples_values)]}, Accuracy: {acc:.2f}")

print("\nBlobs Dataset:")
for i, acc in enumerate(accuracies_dbscan_blobs):
    print(f"Eps: {eps_values[i // len(min_samples_values)]}, Min Samples: {min_samples_values[i % len(min_samples_values)]}, Accuracy: {acc:.2f}")
```

```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Parametry dla k-means
k_values_kmeans = [2, 3, 4]
kmeans_algorithms = [KMeans(n_clusters=k, random_state=42) for k in k_values_kmeans]

# Eksperymenty na zbiorze "circles"
circles_data = [circles] * len(k_values_kmeans)
circles_titles = [f'KMeans on Circles - {k} Clusters' for k in k_values_kmeans]

accuracies_circles = []

for alg, data, title in zip(kmeans_algorithms, circles_data, circles_titles):
    labels = alg.fit_predict(data)
    
    # Obliczanie Silhouette Score dla oceny jakości podziału na klastry
    accuracy = silhouette_score(data, labels)
    title = f'{title} - Silhouette Score: {accuracy:.2f}'
    accuracies_circles.append(accuracy)
    
    plot_clusters(data, labels, title)

# Eksperymenty na zbiorze "moons"
moons_data = [moons] * len(k_values_kmeans)
moons_titles = [f'KMeans on Moons - {k} Clusters' for k in k_values_kmeans]

accuracies_moons = []

for alg, data, title in zip(kmeans_algorithms, moons_data, moons_titles):
    labels = alg.fit_predict(data)
    
    # Obliczanie Silhouette Score dla oceny jakości podziału na klastry
    accuracy = silhouette_score(data, labels)
    title = f'{title} - Silhouette Score: {accuracy:.2f}'
    accuracies_moons.append(accuracy)
    
    plot_clusters(data, labels, title)

# Eksperymenty na zbiorze "ellipses"
ellipses_data = [ellipses] * len(k_values_kmeans)
ellipses_titles = [f'KMeans on Ellipses - {k} Clusters' for k in k_values_kmeans]

accuracies_ellipses = []

for alg, data, title in zip(kmeans_algorithms, ellipses_data, ellipses_titles):
    labels = alg.fit_predict(data)
    
    # Obliczanie Silhouette Score dla oceny jakości podziału na klastry
    accuracy = silhouette_score(data, labels)
    title = f'{title} - Silhouette Score: {accuracy:.2f}'
    accuracies_ellipses.append(accuracy)
    
    plot_clusters(data, labels, title)
```
